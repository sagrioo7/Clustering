```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# Import Dataset
airline=pd.read_csv('https://raw.githubusercontent.com/TejasSutar01/AIRLINES_H_CLUSTERING/master/EastWestAirlines.csv')
airline
```

```python
airline.info()
```

```python
airline2=airline.drop(['ID#'],axis=1)
airline2
```

```python
# Normalize heterogenous numerical data 
airline2_norm=pd.DataFrame(normalize(airline2),columns=airline2.columns)
airline2_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(airline2_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
airline2['clustersid']=hclusters.labels_
airline2
```

```python
airline2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airline2['clustersid'],airline2['Balance'], c=hclusters.labels_) 
```

```python
# Kmeans
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crim1=pd.read_csv('crime_data.csv')
crim1
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
df_norm=norm_func(crim.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crim1['Clust']=x
crim1
```

```python
crim1.iloc[:,1:5].groupby(crim1.Clust).mean()
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
dbscan.labels_
```
